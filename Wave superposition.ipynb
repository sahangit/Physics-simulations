# Creating beats

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# funciton
# angular displacement and frequence

def w(f):
    return 2*np.pi*f

def wave(w):
    return np.sin(w*x)
```

```python
x = np.linspace(0, 1 ,400)

fig, axs = plt.subplots(3, sharex=True, sharey=True)
fig.suptitle('Sharing both axes')
axs[0].plot(x, wave(w(40)))
axs[1].plot(x, wave(w(37)))
axs[2].plot(x, wave(w(40))+wave(w(37)))
```

# Creating standing waves

## Standing waves on a stretched string

```python

import numpy as np
from matplotlib import pyplot as plt
from matplotlib import animation
plt.rcParams['animation.ffmpeg_path'] = '/opt/local/bin/ffmpeg'

fig = plt.figure()
ax = plt.axes(xlim=(0, 2), ylim=(-2, 2))
line, = ax.plot([], [], lw=2)

def init():
    line.set_data([], [])
    return line,

def animate(i):
    x = np.linspace(0, 2, 1000)
    y = np.sin(2 * np.pi * (x - 0.01 * i))
    line.set_data(x, y)
    return line,

anim = animation.FuncAnimation(fig, animate, init_func=init, frames=200, interval=20, blit=True)

writergif = animation.PillowWriter(fps=30) 
anim.save(wave anwriter=writergif)
```
